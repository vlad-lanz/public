```python
def remove_after_stick(my_str: str) -> str:
    '''Функции убирает из строки всё, начиная с палочки включительно'''
    stick_index = my_str.find('|')
    return my_str[:stick_index].replace(" ", '')
```

```python
def add_coma(my_str: str) -> str:
    '''Функция убирает в строке пробелы и добавляет запятую по индексу -2'''
    my_str = my_str.replace(" ", '')
    return f"{my_str[:-2]},{my_str[-2:]}"
```

```python
def update_df(df: pd.DataFrame) -> pd.DataFrame:
    '''Функция обрезает dataframe срезом, чтобы он остался dataframe, а не превратился в 
       numpy.series, а также ставит имя столбика, чтобы всё склеилось'''
    df = df.iloc[:, 3:4]  # здесь для этого сделан срез, а не [:, 3]
    df.columns = ['Значение']
    return df
```

```python
# результат работы - excel файлики
for id in range(1, 2300):
    try:
        pd.read_html(f'https://monitoring.miccedu.ru/iam/{2023}/_vpo/inst.php?id={id}')
        
        university_name = pd.read_html(f'https://monitoring.miccedu.ru/iam/{2023}/_vpo/inst.php?id={id}')[2].iloc[:, -1][0]

        university_df = pd.DataFrame()
        
        for year in range(2015, 2024):
            # если цикл не проходит, то делаем continue
            try:
                # здесь хранится список со всеми таблицами
                all_tables = pd.read_html(f'https://monitoring.miccedu.ru/iam/{year}/_vpo/inst.php?id={id}')
        
                if year == 2015:  # в зависимости от года берётся свой список с индексами таблиц
                    table_list: list = id_year_dict['15']
                elif 2016 <= year <= 2018:
                    table_list: list = id_year_dict['16-18']
                elif year >= 2019:
                    table_list: list = id_year_dict['19+']
        
                # основные показатели
                if year >= 2020:
                    general = all_tables[table_list[0]].iloc[:, 5:6]
                    general.iloc[:, 0] = general.iloc[:, 0].apply(remove_after_stick)
                    general.columns = general.columns.droplevel()
                elif year <= 2019:
                    if year == 2019:
                        general = all_tables[table_list[0]].iloc[1:, 2:3]
                    elif year <= 2018:
                        general = all_tables[table_list[0]].iloc[1:, 2:3].drop(6, axis=0)  # если год меньше 2018 включительно, то нужно удалить один показатель
                    general["Значение показателя"] = general["Значение показателя"].str.replace(" ", '')
                    general.iloc[:3, 0] = general.iloc[:3, 0].apply(add_coma)
                    general.iloc[4:, 0] = general.iloc[4:, 0].apply(add_coma)  
                general.columns = ['Значение']
        
                # образование - дальше всё по одной схеме
                education = all_tables[table_list[1]].iloc[1:, :]              # здесь всегда беру индекс необходимой таблицы
                education[3] = education[3].str.replace(" ", '')               # везде потом удаляю пробелы
                education.iloc[:4, 3] = education.iloc[:4, 3].apply(add_coma)  # добавляю запятые в нужные колонки
                education.iloc[7:, 3] = education.iloc[7:, 3].apply(add_coma)
                education = update_df(education)
        
                # научно-исследовательская
                science = all_tables[table_list[2]].iloc[1:, :]
                science[3] = science[3].str.replace(" ", '')
                science.iloc[:10, 3] = science.iloc[:10, 3].apply(add_coma)
                science.iloc[11:14, 3] = science.iloc[11:14, 3].apply(add_coma)
                science.iloc[15:16, 3] = science.iloc[15:16, 3].apply(add_coma)
                science = update_df(science)
        
                # международный
                international = all_tables[table_list[3]].iloc[1:, :]
                international[3] = international[3].str.replace(" ", '')
                international.iloc[:8, 3] = international.iloc[:8, 3].apply(add_coma)
                international.iloc[9:, 3] = international.iloc[9:, 3].apply(add_coma)
                international = update_df(international)
        
                # финансово-экономический блок
                finance = all_tables[table_list[4]].iloc[1:, :]
                finance[3] = finance[3].str.replace(" ", '')
                finance.iloc[:4, 3] = finance.iloc[:4, 3].apply(add_coma)
                finance = update_df(finance)
        
                # инфраструктура
                infrastructure = all_tables[table_list[5]].iloc[1:, :]
                infrastructure[3] = infrastructure[3].str.replace(" ", '').apply(add_coma)
                infrastructure = update_df(infrastructure)
        
                # кадры
                people = all_tables[table_list[6]].iloc[1:, :]
                people[3] = people[3].str.replace(" ", '').apply(add_coma)
                people = update_df(people)
        
                # дополнительное: обучение
                edu_add = all_tables[table_list[7]].iloc[1:13, :].droplevel(level=1, axis=1)
                edu_add.iloc[:, 3] = edu_add.iloc[:, 3].str.replace(" ", '')
                edu_add.iloc[4:8, 3] = edu_add.iloc[4:8, 3].apply(add_coma)
                edu_add = update_df(edu_add)
        
                # дополнительное: научная деятельность
                science_add = all_tables[table_list[7]].iloc[14:25, :].droplevel(level=1, axis=1)
                science_add.iloc[:, 3] = science_add.iloc[:, 3].str.replace(" ", '')
                science_add.iloc[2:3, 3] = science_add.iloc[2:3, 3].apply(add_coma)
                science_add.iloc[8:9, 3] = science_add.iloc[8:9, 3].apply(add_coma)
                science_add = update_df(science_add)
        
                # дополнительное: кадры
                people_add = all_tables[table_list[7]].iloc[26:35, :].droplevel(level=1, axis=1)
                people_add.iloc[:, 3] = people_add.iloc[:, 3].str.replace(" ", '')
                people_add.iloc[3:, 3] = people_add.iloc[3:, 3].apply(add_coma)
                people_add = update_df(people_add)
        
                # дополнительные: международная деятельность
                international_add = all_tables[table_list[7]].iloc[36:45, :].droplevel(level=1, axis=1)
                international_add.iloc[:, 3] = international_add.iloc[:, 3].str.replace(" ", '')
                international_add.iloc[1:2, 3] = international_add.iloc[1:2, 3].apply(add_coma)
                international_add.iloc[6:8, 3] = international_add.iloc[6:7, 3].apply(add_coma)
                international_add = update_df(international_add)
        
                # дополнительные: инфраструктура
                infrastructure_add = all_tables[table_list[7]].iloc[46:55, :].droplevel(level=1, axis=1)
                infrastructure_add.iloc[:, 3] = infrastructure_add.iloc[:, 3].str.replace(" ", '')
                infrastructure_add.iloc[5:6, 3] = infrastructure_add.iloc[5:6, 3].apply(add_coma)
                infrastructure_add.iloc[7:8, 3] = infrastructure_add.iloc[7:8, 3].apply(add_coma)
                infrastructure_add = update_df(infrastructure_add)
        
                # дополнительные: финансы
                finance_add = all_tables[table_list[7]].iloc[56:, :].droplevel(level=1, axis=1)
                finance_add.iloc[:, 3] = finance_add.iloc[:, 3].str.replace(" ", '')
                finance_add.iloc[2:, 3] = finance_add.iloc[2:, 3].apply(add_coma)
                finance_add = update_df(finance_add)
        
                # действия над финальной таблицей: склеиваю всё
                concatted = pd.DataFrame(pd.concat([general, education, science, international, finance, infrastructure,
                                                    people, edu_add, science_add, people_add, international_add,
                                                    infrastructure_add, finance_add], axis=0))
                concatted.index = dict_with_formats.keys()  # нужно проставлять здесь, чтобы конкатенация сработала
                concatted = concatted.transpose()
                concatted['Год'], concatted['Университет'] = year - 1, university_name  # year - 1, потому что мониторинг строится по предыдущему году
                
                university_df = pd.concat([university_df, concatted], axis=0)
        
                sleep(0.5)
            except:
                print(f"По университету: {university_name}, id: {id}, год: {year}, ошибка в цикле")
                continue

        # проходим универ и делаем выгрузку
        file_name = f'C:/python_work/выгрузки/1_мониторинг/full_backup/{id}.xlsx'
        university_df.to_excel(file_name, index=None)
    
    except ValueError:
        print(f"Университета с id {id} не существует")
    except IndexError:
        print(f"По университету: {university_name} с id: {id}, по году {year}, ошибка IndexError")
```
