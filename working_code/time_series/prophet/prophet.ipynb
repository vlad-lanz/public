```python
import prophet
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monthly-car-sales.csv')
```

```python
df.head()
```

```python
model = prophet.Prophet()
```

```python
df['ds'] = pd.to_datetime(df['Month'])
df['y'] = df['Sales']
```

```python
df = df.drop(['Month', 'Sales'], axis=1)
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(2000)
```

```python
forecast = model.predict(future)
```

```python
plt.plot()
```

```python
model.plot(forecast)
```

```python
model.plot_components(forecast)
```
