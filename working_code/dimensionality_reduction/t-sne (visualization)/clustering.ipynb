```python
import pandas as pd

from sklearn.manifold import TSNE

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.data.txt', sep=r'\s+', header=None)
```

```python
x = df.iloc[:, :-1]
```

<h1>t-sne импользуется для визуализации, связь между переменными как в PCA теряется</h1>
<h2>Основные параметры:
<li>n-components - какая размерность, для изображения на 2d графиках использовать 2, для 3d - 3 (возможно)</li>
<li>perplexity - чем выше, тем больше график по форме приближается к букве S</li>
</h2>

```python
tsne = TSNE(n_components=2, perplexity=100, random_state=42)
X_reduced_tsne = tsne.fit_transform(x)
```

```python
first = X_reduced_tsne[:, 0]
second = X_reduced_tsne[:, 1]
```

```python
visual = pd.DataFrame({'first': first,
              'second': second})
```

```python
for plex in range(10, 201, 10):
    tsne = TSNE(n_components=2, perplexity=plex, random_state=42)
    X_reduced_tsne = tsne.fit_transform(x)
    first = X_reduced_tsne[:, 0]
    second = X_reduced_tsne[:, 1]
    visual = pd.DataFrame({'first': first,
                  'second': second})
    sns.lmplot(data=visual,
                x='first',
                y='second')
    plt.title(f"\n\nPerplexity = {plex}")
```
