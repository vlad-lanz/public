```python
import numpy as np
import plotly.express as px
import seaborn as sns
```

<h1><center>Сравним распределения 4.05 звезды с 100 оценками против 4 звезды с 250 оценками</center></h1>

<h1>Начнём с 4.05 звезд 100 оценок, каждый раз будем симулировать 10_000 операций</h1>

```python
# допустим, это распределение 4 звёзд
distribution_1 = np.array([0.05, 0.05, 0.05, 0.5, 0.35])
```

```python
ratings = np.arange(1, 6)
```

```python
d1_simulation = np.random.choice(
    a=ratings,
    p=distribution_1,
    size=(10_000, 100))
```

```python
d1_avg = np.array([np.mean(elem) for elem in d1_simulation])
```

```python
np.mean(d1_avg)
```

```python
print(f'''Среднее: {np.mean(d1_avg)}
Стандартное отклонение {np.std(d1_avg)}''')
```

```python
sns.kdeplot(d1_avg)
```

<h1>Теперь рассмотрим 250 оценок с 4 звёздами</h1>

```python
distribution_2 = np.array([0.05, 0.05, 0.05, 0.55, 0.3])
```

```python
d2_simulation = np.random.choice(
    a=ratings,
    p=distribution_2,
    size=(10_000, 250)
)
```

```python
d2_avg = np.array([np.mean(elem) for elem in d2_simulation])
```

```python
print(f'''Среднее: {np.mean(d2_avg)}
Стандартное отклонение: {np.std(d2_avg)}''')
```

```python
sns.kdeplot(d2_avg)
sns.kdeplot(d1_avg)
```

<h2>В 1 распределении среднее оказалось выше, поэтому из 1 вычитаем второе</h2>

```python
sns.kdeplot(d1_avg - d2_avg)
```

```python
# разница 2 распределений
difference = d1_avg - d2_avg
```

```python
sum(difference > 0) / difference.shape[0] * 100
```

<h2>То есть в 67% случаев среднее в 1 распределении оказывалось выше, чем во втором распределении</h2>
